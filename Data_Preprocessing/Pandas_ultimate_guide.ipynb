## Pandas Sampling:

```
import pandas as pd
from sklearn.datasets import load_iris
iris_df=load_iris()
df_iris=pd.DataFrame(iris_df.data,columns=iris_df.feature_names)
df_iris['Target']=iris_df.target
df_iris['Target']=df_iris['Target'].apply(lambda x:iris_df.target_names[x])
```

```
df_iris.sample(n=3,random_state=1)
```

```
def getting_sample_data(df):
    return df.sample(n=3,random_state=1)
df=df_iris.groupby('Target').apply(getting_sample_data)
df
```

##Crosstab

```
import pandas as pd
from sklearn.datasets import load_iris
iris_df=load_iris()
df_iris=pd.DataFrame(iris_df.data,columns=iris_df.feature_names)
df_iris['Target']=iris_df.target
df_iris['Target']=df_iris['Target'].apply(lambda x:iris_df.target_names[x])
df_iris.head(3)
```

```
labels = range(1,5)
df_iris['M_Group'] = pd.qcut(df_iris['sepal length (cm)'], q=4, labels=labels)
pd.crosstab(df_iris['M_Group'],df_iris['Target'],margins=True)
```

##Join

```
df_sepal_len = pd.DataFrame(df_iris.groupby('Target')['sepal length (cm)'].max())
df_sepal_wid = pd.DataFrame(df_iris.groupby('Target')['sepal width (cm)'].max())
df_petal_len = pd.DataFrame(df_iris.groupby('Target')['petal length (cm)'].max())
df_petal_wid = pd.DataFrame(df_iris.groupby('Target')['petal width (cm)'].max())
```

```
df_sepal_len
```

```
df_01 = df_sepal_len.join(df_sepal_wid,on=None,how='inner')
df_01 =df_01.join(df_petal_len,on=None,how='inner')
df_01 =df_01.join(df_petal_wid,on=None,how='inner')
df_01
```

```
df_01 =df_sepal_len.merge(df_sepal_wid,left_index=True,right_index=True,how='inner')
```
